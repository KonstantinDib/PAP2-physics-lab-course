```python
#Importieren der benoetigten Pakete
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
import numpy as np
from scipy.optimize import curve_fit 

sa=1e-3*np.array([0.000,0.040,0.800,1.201,1.600])
se=1e-3*np.array([2.975,3.015,3.973,4.173,4.572])
sa_err=0.009e-3*np.ones(5)
se_err=0.009e-3*np.ones(5)
m=np.array([11138,11153,11563,11343,11177])
m_err=50*np.ones(5)

lam=2*(se-sa)/m
lam_sys=(4/(m**2)*(sa_err**2+se_err**2)+(2*(se-sa)*m_err/(m**2))**2)**0.5

lam_mean=np.mean(lam)
lam_mean_sys=(np.sum(lam_sys**2)/len(lam_sys))**0.5
lam_mean_std=np.std(lam)/np.sqrt(len(lam))

print(lam)
print(np.mean(lam))
print(lam_sys)
print(lam_mean_sys)
print(lam_mean_std)

lam_lit=532e-9
lam_lit_err=1e-9

diff_1=np.abs(lam_mean-lam_lit)
diff_1_err=(lam_mean_sys**2+lam_lit_err**2)**0.5

diff_2=np.abs(lam_mean-lam_lit)
diff_2_err=(lam_mean_std**2+lam_lit_err**2)**0.5

print("Differenz der beiden Werte mit systematischem Fehler: "+str(diff_1)+" +/- "+str(diff_1_err))
print("Differenz der beiden Werte mit statistischem Fehler: "+str(diff_2)+" +/- "+str(diff_2_err))
print("Also beides 1 simga")
#print(np.sum(sa)*1e3)
```

```python
T_0=273.15
p_0=760     #101325
T=23.8+273.15
T_err=0.2
a=50e-3
a_err=0.05e-3

del_m=np.array([0,5,10,15,20,25,30,35,40,45,50])
p_1=-1*np.array([750,675,600,525,450,375,300,225,150,75,0])
p_2=-1*np.array([750,665,570,500,425,350,270,185,110,35])
p_3=-1*np.array([750,675,605,530,455,360,280,200,125,50])

plt.grid(ls=":")
plt.xlabel("Druck $p$ [Torr]",size=15)
plt.ylabel("Anzahl der Ringe $\Delta m$",size=15)
plt.xlim(-800,50)
plt.ylim(-5,55)
#plt.title("Auftragen der Anzahl der Ringe $\Delta m$ gegen den Druck $p$", size=15)

def linear(x,a,b):
    return a*x+b

popt_1,pcov_1=curve_fit(linear,p_1,del_m)
popt_2,pcov_2=curve_fit(linear,p_2,del_m[:-1])
popt_3,pcov_3=curve_fit(linear,p_3,del_m[:-1])

plt.plot(p_1,linear(p_1,*popt_1),color="red")#,linestyle='--')
plt.plot(p_2,linear(p_2,*popt_2),color="green")#,linestyle='--')
plt.plot(p_3,linear(p_3,*popt_3),color="blue",linestyle='--')

plt.plot(p_1,del_m,marker='o',color="red",linewidth=0)
plt.plot(p_2,del_m[:-1],marker='.',color="green",linewidth=0)
plt.plot(p_3,del_m[:-1],marker='x',color="blue",linewidth=0)
plt.savefig('figures/Michelson.pdf', format='PDF')

#plt.text(8,335, "Funktion: y=a*x+b", fontsize=12)
#plt.text(8,300, "a=("+str(round(popt[0],3))+"$\pm$"+str(round(pcov[0,0]**0.5,3))+")$\mu m^2/s$", fontsize=12)
#plt.text(8,265, "b=("+str(round(popt[1],3))+"$\pm$"+str(round(pcov[1,1]**0.5,3))+")$\mu m^2/s$", fontsize=12)

n_0_1=lam_lit*p_0*T/(2*a*T_0)*popt_1[0]+1
n_0_2=lam_lit*p_0*T/(2*a*T_0)*popt_2[0]+1
n_0_3=lam_lit*p_0*T/(2*a*T_0)*popt_3[0]+1

popt_1_err=0
popt_2_err=0
popt_3_err=0

n_0_1_err=(n_0_1-1)*np.sqrt((lam_lit_err/lam_lit)**2+(a_err/a)**2+(T_err/T)**2+(popt_1_err/popt_1[0])**2)
#print((lam_lit_err/lam_lit)**2,(a_err/a)**2,(T_err/T)**2)
n_0_2_err=(n_0_2-1)*np.sqrt((lam_lit_err/lam_lit)**2+(a_err/a)**2+(T_err/T)**2+(popt_2_err/popt_2[0])**2)
n_0_3_err=(n_0_3-1)*np.sqrt((lam_lit_err/lam_lit)**2+(a_err/a)**2+(T_err/T)**2+(popt_3_err/popt_3[0])**2)

print(n_0_1)
print(n_0_2)
print(n_0_3)

print(n_0_1_err)
print(n_0_2_err)
print(n_0_3_err)

n_0_lit=1.00028

diff_1=np.abs(n_0_1-n_0_lit)
diff_2=np.abs(n_0_2-n_0_lit)
diff_3=np.abs(n_0_3-n_0_lit)

print("Differenz des ersten Messwertes mit dem Literaturwert: "+str(diff_1)+" +/- "+str(n_0_1_err))
print("Differenz des zweiten Messwertes mit dem Literaturwert: "+str(diff_2)+" +/- "+str(n_0_2_err))
print("Differenz des dritten Messwertes mit dem Literaturwert: "+str(diff_3)+" +/- "+str(n_0_3_err))
```
