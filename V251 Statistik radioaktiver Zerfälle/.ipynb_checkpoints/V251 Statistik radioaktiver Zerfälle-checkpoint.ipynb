# V251: Statistik radioaktiver Zerfälle

## Auswertung der Messergebnisse

```python
#Benötigte Pakete
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.optimize import curve_fit
from scipy.stats import chi2
from scipy.special import gamma

plt.rcParams["font.family"]='serif'
plt.rcParams["figure.figsize"][0] = 14
plt.rcParams["figure.figsize"][1] = 9
plt.rcParams['errorbar.capsize']=2
```

### Teil 1: Messung der Zählrohrcharakteristik

```python
#Zählrohrspannung
U0=np.linspace(420,570,7)
U0_err=5*np.ones(7)

#Zerfälle
decays=np.array([2349,2677,2688,2661,2638,2720,2788])
decays_err=np.sqrt(decays)
print(decays_err.tolist())
```

```python
#Fit des Plateaubereichs

def linear(x,a,b):
    return a*x+b

popt,pcov=curve_fit(linear,U0[1:],decays[1:],sigma=decays_err[1:])
```

```python
#Plot der Messdaten: U_ind gegen omega
plt.errorbar(U0, decays, xerr=U0_err,yerr=decays_err,linewidth=1,linestyle='',marker='.',color='darkblue', label='Messwert mit Fehler')
plt.plot(np.linspace(430,590,2),linear(np.linspace(430,590,2),*popt),linewidth=0.5,color='darkred',label='Linearer Fit: $m_{fit} = ($' + str(round(popt[0],1)) + ' $\pm$ ' + str(round(pcov[0,0]**0.5,1)) + r') $\frac{1}{V}$')
plt.xlabel(r'Zählrohrspannung $ U_0$  [V]', size=20)
plt.ylabel(r'Anzahl der Ereignisse N', size=20)
plt.title(r'Diagramm 1: Zählrohrcharakteristik', size=22)
plt.grid(ls='dotted')

plt.legend(frameon=True,fontsize='large', borderpad=1.5, borderaxespad=1.5)

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V251Diagramm1.pdf', format='PDF')
```

### Teil 2: Plateaubereich des Zählrohrs

```python
#Messdauer 60s
N0=np.array([13450,24302])
N0_err=np.sqrt(N0)

#Messdauer 180s
N1=np.array([13535,24751])
N1_err=np.sqrt(N1)

#Anstieg
rise=N1-N0
rise_err=np.sqrt(N0_err**2+N1_err**2)

#prozentualer Anstieg
percental_rise=rise/N0
percental_rise_err=percental_rise*np.sqrt((N1_err/N1)**2+(N0_err/N0)**2)

print('Der prozentuale Anstieg der Zählrate wurde bestimmt auf:')
print('1. Messreihe: ' + str(percental_rise[0]) + ' +/- ' + str(percental_rise_err[0]))
print('2. Messreihe: ' + str(percental_rise[1]) + ' +/- ' + str(percental_rise_err[1]))
```

```python
#Überprüfe auf Signifikanz
diff_pr=np.abs(percental_rise[0]-percental_rise[1])
diff_pr_err=np.sqrt(percental_rise_err[0]**2+percental_rise_err[1]**2)

print('Wir erhalten für die Differenz der Ergebnisse aus beiden Messreihen:')
print('∆pr = (' + str(diff_pr) + ' +/- ' + str(diff_pr_err))
print()
print('Damit sind die Ergebnisse signifikant verschieden!')
```

### Teil 3: Auswertung der Daten mit hoher mittlerer Ereigniszahl

```python
#Einlesen der Messdaten
quantity,rate=np.loadtxt('Messdaten/V251Messdaten.dat',unpack=True)
rate_err=np.sqrt(rate)

plt.errorbar(quantity, rate, yerr=error,linewidth=1,linestyle='',marker='.',color='darkblue', label='Messwert mit Fehler')
plt.xlabel(r'Anzahl der Zerfälle pro Zeiteinheit [$\frac{1}{s}$]', size=20)
plt.ylabel(r'Häufigkeit', size=20)
plt.title(r'Diagramm 2: Statistik des radioaktiven Zerfalls (hohe mittlere Ereigniszahl)', size=20)
plt.grid(ls='dotted')

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V251Diagramm2.pdf', format='PDF')
```

```python
#Fit der Messdaten
def gaussian(x,A,mu,sig): #mit A: Fläche der Gaussfunktion
    return A/(np.sqrt(2*np.pi)*sig)*np.exp(-(x-mu)**2/2/sig**2)

p=[2000,72,8]
popt2,pcov2=curve_fit(gaussian,quantity[6:-8],rate[6:-8], sigma=rate_err[6:-8],p0=p)
```

```python
#Plot der Messdaten mit Fit
plt.errorbar(quantity, rate, yerr=error,linewidth=1,linestyle='',marker='.',color='darkblue', label='Messwerte mit Fehler')
plt.plot(np.linspace(40,110,500),gaussian(np.linspace(40,110,500),*popt2),linewidth=0.5,color='darkred',label='Gauß-Fit')
plt.xlabel(r'Anzahl der Zerfälle pro Zeiteinheit [$\frac{1}{s}$]', size=20)
plt.ylabel(r'Häufigkeit', size=20)
plt.title(r'Diagramm 3: Statistik des radioaktiven Zerfalls (hohe mittlere Ereigniszahl) mit Gauß-Fit', size=19)
plt.grid(ls='dotted')
plt.legend(frameon=True,fontsize='large',loc='best', borderpad=1.5, borderaxespad=1.5)

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V251Diagramm3.pdf', format='PDF')
```

```python
#Güte des Gauß-Fits
chi_squared=np.sum((gaussian(quantity[6:-8],*popt2)-rate[6:-8])**2/rate_err[6:-8]**2)

#Freiheitsgrade
dof=len(quantity[6:-8])-3

chi_squared_red=chi_squared/dof

print('Wir erhalten die nachfolgenden Werte für die Güte des Fits:')
print('chi_squared= ' + str(chi_squared))
print('chi_squared_red= ' + str(chi_squared_red))
print()

#Fitwahrscheinlichkeit
prob=round(1-chi2.cdf(chi_squared,dof),2)*100

print('Die Fitwahrscheinlichkeit beträgt: ' + str(prob) + ' %')
```

```python
#Vergleich mit Poisson-Verteilung
def poisson(x,A,mu_p):
    return A*np.exp(-mu_p)*mu_p**x/gamma(x+1)

p2=[2000,72]
popt3,pcov3=curve_fit(poisson,quantity[6:-8],rate[6:-8], sigma=rate_err[6:-8], p0=p2)
```

```python
#Plot der Messdaten mit beiden Fits
plt.errorbar(quantity, rate, yerr=error,linewidth=1,linestyle='',marker='.',color='darkblue', label='Messwerte mit Fehler')
plt.plot(np.linspace(40,110,500),gaussian(np.linspace(40,110,500),*popt2),linewidth=0.8,color='darkred',label='Gauß-Fit')
plt.plot(np.linspace(40,110,500),poisson(np.linspace(40,110,500),*popt3),linewidth=0.8,color='darkgreen',linestyle='--',label='Poisson-Fit')
plt.xlabel(r'Anzahl der Zerfälle pro Zeiteinheit [$\frac{1}{s}$]', size=20)
plt.ylabel(r'Häufigkeit', size=20)
plt.title(r'Diagramm 4: Statistik des radioaktiven Zerfalls (hohe mittlere Ereigniszahl) mit Gauß- und Poisson-Verteilung', size=15)
plt.grid(ls='dotted')
plt.legend(frameon=True,fontsize='large',loc='best', borderpad=1.5, borderaxespad=1.5)

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V251Diagramm4.pdf', format='PDF')
```

```python
#Güte des Poisson-Fits
chi_squared2=np.sum((poisson(quantity[6:-8],*popt3)-rate[6:-8])**2/rate_err[6:-8]**2)

#Freiheitsgrade
dof2=len(quantity[6:-8])-2

chi_squared_red2=chi_squared2/dof2

print('Wir erhalten die nachfolgenden Werte für die Güte des Fits:')
print('chi_squared= ' + str(chi_squared2))
print('chi_squared_red= ' + str(chi_squared_red2))
print()

#Fitwahrscheinlichkeit
prob2=round(1-chi2.cdf(chi_squared2,dof2),2)*100

print('Die Fitwahrscheinlichkeit beträgt: ' + str(prob2) + ' %')
```

```python
#Ausgeben der ermittelten Fit-Parameter
print('Für den Gauß-Fit wurde ermittelt:')
print('A = (' + str(popt2[0]) + ' +/- ' + str(pcov2[0,0]**0.5) + ') 1/s')
print('mu = (' + str(popt2[1]) + ' +/- ' + str(pcov2[1,1]**0.5) + ') 1/s')
print('sigma = (' + str(popt2[2]) + ' +/- ' + str(pcov2[2,2]**0.5) + ') 1/s')

print()
print('Für den Poisson-Fit wurde ermittelt:')
print('A = (' + str(popt3[0]) + ' +/- ' + str(pcov3[0,0]**0.5) + ') 1/s')
print('mu = (' + str(popt3[1]) + ' +/- ' + str(pcov3[1,1]**0.5) + ') 1/s')
```

### Teil 4: Auswertung der Daten mit kleiner mittlerer Ereigniszahl
