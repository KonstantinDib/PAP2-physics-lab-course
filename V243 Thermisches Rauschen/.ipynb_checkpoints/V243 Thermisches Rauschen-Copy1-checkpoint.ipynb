## Auswertung der Messergebnisse

```python
#Benötigte Pakete
import numpy as np
from scipy.optimize import curve_fit
import scipy.integrate as integrate
from scipy.stats import chi2
import matplotlib.pyplot as plt
%matplotlib inline 

plt.rcParams['errorbar.capsize']=2
```

### Teil 3: Messung des Frequenzgangs des Verstärkers und des Bandfilters

```python
#Importieren der Messdaten
freq, U_aus=np.loadtxt('E:/Frequenzgang.txt', skiprows=1, usecols=(0,1), unpack=True)

#Abschwächung des Dämpfungsglieds
D=1e-3

#Eingangsspannung
U_ein=0.2
#Frequenzgang g(freq)
g=U_aus/(U_ein*D)
```

```python
#Plot der Messdaten
plt.loglog(freq,g,linestyle='None', marker='.')
plt.axis([4E2,1.5e5,10,1.5E3])
plt.xlabel(r'Frequenz $\nu$ [Hz]',size=13)
plt.ylabel(r'g($\nu$)',size=13)
plt.title('Diagramm 5: Gemessener Frequenzgang',size=13)


#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('E:/V243Diagramm5.pdf', format='PDF')
```

```python
#Plot der Messdaten ohne störende Randwerte

plt.loglog(freq[17:-43],g[17:-43],linestyle='None', marker='.')
plt.axis([4E2,1.5e5,10,1.5E3])
plt.xlabel(r'Frequenz $\nu$ [Hz]',size=13)
plt.ylabel(r'g($\nu$)',size=13)
plt.title('Diagramm 6: Gemessener Frequenzgang (korrigiert)',size=13)

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('E:/V243Diagramm6.pdf', format='PDF')
```

```python
#Fit der theoretischen Kurve an unsere Messdaten
def fit_function(freq,V,W1,W2,n1,n2):
    return V/(np.sqrt(1+1/(freq/W1)**(2*n1))*np.sqrt(1+(freq/W2)**(2*n2)))

#Startwerte:
#Verstärkung V=1000, untere Grenzfrequenz W1= 1000, obere Grenzfrequenz W2= 50000

p0= [1000, 1000, 50000, 5, 5]

popt1,pcov1= curve_fit(fit_function, freq[17:-43],g[17:-43], p0)
```

```python
#Plot der Messdaten inklusive Fit 

plt.loglog(freq[17:-43],g[17:-43],linestyle='None', marker='.', label='Messdaten')
plt.loglog(freq,fit_function(freq,*popt1), label='Fit der Messdaten')
plt.axis([4E2,1.5e5,10,1.5E3])
plt.xlabel(r'Frequenz $\nu$ [Hz]',size=13)
plt.ylabel(r'g($\nu$)',size=13)
plt.title('Diagramm 7: Gemessener Frequenzgang mit zugehöriger Fitkurve',size=13)
plt.legend(loc='best')

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('E:/V243Diagramm7.pdf', format='PDF')
```

#### Numerische Integration

```python
def fit_function_squared(freq,V,W1,W2,n1,n2):
    return fit_function(freq,V,W1,W2,n1,n2)**2

B=integrate.quad(fit_function_squared, freq[17], freq[-43], args=tuple(popt1))

print('Das Integral liefert den Wert: {value:.4e}'.format(value=B[0]))
```

### Teil 2: Bestimmung der Boltzmannkonstante

```python
#Widerstand und gemessene Rauschspannung

#Achtung: unsere Messwerte einsetzen

R=np.array([5e3, 10e3, 15e3, 20e3, 25e3, 30e3])

U_aus=np.array([2.4137e-3, 3.1331e-3, 3.7152e-3, 4.2273e-3, 4.6859e-3, 5.1074e-3 ])
U_aus_err=np.array([0.0009e-3, 0.0011e-3, 0.0018e-3, 0.0017e-3, 0.0018e-3, 0.0023e-3])

U_V=1.3579e-3
U_v_err=0.0006e-3

diff_U=U_aus**2-U_V**2
diff_U_err=np.sqrt((2*U_aus*U_aus_err)**2+(2*U_V*U_v_err)**2)
```

```python
#linearer Fit
def linear(x,c):
    return c*x

popt2,pcov2= curve_fit(linear,R,diff_U, 0)

#Plot der Messdaten mit linearem Fit
x0=np.linspace(0,32000,1000)
plt.errorbar(R,diff_U, yerr=diff_U_err, fmt='.', label='Messdaten mit Fehler')
plt.plot(x0,linear(x0,*popt2), linewidth=0.8, label=r'linearer Fit: $c\cdot x$')
plt.axis([0,3.2e4,0,2.5e-5])
plt.xlabel(r'Widerstand $R$ [$\Omega$]',size=13)
plt.ylabel(r'$∆U^2=(U_{aus}^2-U_V^2)$ [$V^2$]',size=13)
plt.title('Diagramm 8: Bestimmung der Boltzmannkonstanten',size=13)
plt.yticks([0,5e-6,1e-5,1.5e-5,2e-5,2.5e-5],[r'$0$',r'$5 \cdot 10^{-6}$',r'$1 \cdot 10^{-5}$',r'$1.5 \cdot 10^{-5}$',r'$2 \cdot 10^{-5}$'])
plt.text(12000,2e-6, "$c$ = ("+str(round(popt2[0]*1e10,3))+"$\pm$"+str(round(pcov2[0,0]**0.5*1e10,3))+")" + ' ' r"$\times 10^{-10}$" + ' ' + r"$\frac{V^2}{\Omega}$",bbox={'facecolor':'white', 'alpha':0.5, 'pad':10},size=13)
plt.legend(loc='best')

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('E:/V243Diagramm8.pdf', format='PDF')
```

```python
#Güte des Fits
chi_squared=np.sum((linear(R,*popt2)-diff_U)**2/diff_U_err**2)

#Freiheitsgrade
dof=5

chi_squared_red=chi_squared/dof

print('Wir erhalten die nachfolgenden Werte für die Güte des Fits:')
print('chi_squared= ' + str(chi_squared))
print('chi_squared_red= ' + str(chi_squared_red))
print()

#Fitwahrscheinlichkeit
prob=round(1-chi2.cdf(chi_squared,dof),2)*100

print('Die Fitwahrscheinlichkeit beträgt: ' + str(prob) + ' %')
```

```python
#Theorie: c=4kTB

T=296.4
T_err=0.5

B_err=B[0]*0.2

k=popt2[0]/(4*B[0]*T)
k_err=k*np.sqrt((pcov2[0,0]**0.5/popt2[0])**2+(B_err/B[0])**2+(T_err/T)**2)
k_std=k*(pcov2/popt2)

print('Aus unseren Messwerten folgt für die Boltzmannkonstante:')
print('k = ' + str(k) + ' +/- ' + str(k_err) + ' +/- ' + str(k_std[0][0]))
```
