## Auswertung Versuch 213 Kreisel

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
plt.style.use('seaborn-white')

x = np.array([0, 120, 240, 360, 480, 600, 720])
x_err=0.1*np.ones(7)
y = np.array([678.2, 624.7, 577.0, 533.7, 493.1, 456.4, 422.7])
y_err = 5 * np.ones(7)

def exponential(x,a,b):
    return a*np.exp(-b*x)

p = (600, 0.0005)

popt,pcov=curve_fit(exponential,x,y,sigma=y_err, p0=p)

plt.rcParams['errorbar.capsize']=2

plt.plot(x, y, marker='.',markersize=5, color='black',linewidth=1,linestyle=' ')
plt.yscale('log')
plt.errorbar(x,y,xerr=x_err,yerr=y_err,linewidth=1,linestyle='',color='black',label='Messpunkte mit Fehler')
plt.plot(x,exponential(x, popt[0], popt[1]),color='red', label= 'exponentieller Fit')
plt.xlabel('Zeit ['+'$s$]', size=15)
plt.ylabel('Frequenz'+ ' '+r'$\nu$' + r' ['+r'$\frac{1}{min}$]' , size=15)
plt.title('Diagramm 1: Bestimmung der Dämpfungskonstanten', size=15)
plt.grid(ls='dotted')
plt.legend(frameon=True,fontsize='large')
plt.ylim(380,730)
plt.text(0,410, "$\delta$ = ("+str(int(popt[1]*1e6+0.5))+"$\pm$"+str(int(pcov[1,1]**0.5*1e6+0.5))+")" + ' ' r"$\times 10^{-6}$" + ' ' + "$s^{-1}$",bbox={'facecolor':'white', 'alpha':0.5, 'pad':10},size=13)


#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V213Diagramm1', format='PDF')

print("a =", popt[0], "+/-", pcov[0,0]**0.5)
print("b =", popt[1], "+/-", pcov[1,1]**0.5)
```
