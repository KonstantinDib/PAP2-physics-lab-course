## Auswertung der Messergebnisse

```python
#Benötigte Pakete
import numpy as np
from scipy.optimize import curve_fit
import scipy.integrate as integrate
from scipy.stats import chi2
import matplotlib.pyplot as plt
%matplotlib inline 

#Plot settings
plt.rcParams["font.family"]='serif'
plt.rcParams["figure.figsize"][0] = 10
plt.rcParams["figure.figsize"][1] = 7
plt.rcParams['errorbar.capsize']=2
```

### Teil 1: Messung des Nulleffekts

```python
#Betriebsspannung
U1=520 #V
U1_err=5 #V

#Radius des Zählrohrs
r=7e-3 #m

#Zerfälle in 5 Minuten = 300s
n0=140
n0_err=np.sqrt(140)

#Normierung auf Zerfälle/Sekunde
n0_s=140/300
n0_s_err=n0_err/300
```

### Teil 2: Absorption von $\beta$-Strahlung in Aluminium

```python
#Kennnummer des Präparats: GS 527

#Anzahl Zerfälle
n_beta=np.array([1574,967,678,443,315,181,125,307,166,122,75,67,359])
n_beta_err=np.sqrt(n_beta)

#Torrzeit
T=np.array([30,30,30,30,30,30,30,120,120,120,120,120,300])

#Absorberdicke in mm
x1=np.array([0,0.3,0.6,0.9,1.2,1.5,1.8,2.1,2.4,2.7,3.0,3.3,4.3])
#fehlerfrei?

########################################
# aus letztem Messwert n0_beta bestimm #
# n0_beta=n_beta[12]                   #
#######################################

#Abzug der Nullmessung unter Berücksichtigung der jeweiligen Torrzeit
n_beta_korr_30=n_beta[:7]-(30*n0_s)*np.ones(7)
n_beta_korr_30_err=np.sqrt(n_beta_err[:7]**2+(30*n0_s_err*np.ones(7))**2)

n_beta_korr_120=n_beta[7:12]-(120*n0_s)*np.ones(5)
n_beta_korr_120_err=np.sqrt(n_beta_err[7:12]**2+(120*n0_s_err*np.ones(5))**2)

n_beta_korr_300=n_beta[12]-n0
n_beta_korr_300_err=np.sqrt(n_beta_err[12]**2+(n0_err**2))

#Zusammenführen der drei Arrays 
n_beta_korr=[]
for i in range(len(n_beta_korr_30)):
    n_beta_korr.append(n_beta_korr_30[i])
for i in range(len(n_beta_korr_120)):
    n_beta_korr.append(n_beta_korr_120[i])
n_beta_korr.append(n_beta_korr_300)

#Analog für die Fehler 
n_beta_korr_err=[]
for i in range(len(n_beta_korr_30_err)):
    n_beta_korr_err.append(n_beta_korr_30_err[i])
for i in range(len(n_beta_korr_120_err)):
    n_beta_korr_err.append(n_beta_korr_120_err[i])
n_beta_korr_err.append(n_beta_korr_300_err)

n_beta_korr=np.array(n_beta_korr)
n_beta_korr_err=np.array(n_beta_korr_err)
```

```python
#lineare Fit-Funktion
def exp_fit(x,a,b):
    return a*np.exp(-b*x)

popt1,pcov1=curve_fit(exp_fit,x1[:7],n_beta_korr[:7], sigma=n_beta_korr_err[:7])
```

```python
#Plot der Messdaten
plt.errorbar(x1,n_beta_korr, yerr=n_beta_korr_err,linewidth=1,linestyle='',marker='.',color='darkblue',label='Messwerte mit Fehler')
plt.plot(np.linspace(-0.2,4.2,2),exp_fit(np.linspace(-0.2,4.2,2),*popt1),color='darkred',label='Linearer Fit', linewidth=0.8)
plt.xlabel(r'Absorberdicke $ x$  [mm]', size=20)
plt.ylabel(r'Anzahl der Zerfälle', size=20)
plt.yscale('log')
plt.legend(frameon=True,fontsize='large',loc='best', borderpad=1.5, borderaxespad=1.5)
plt.title(r'Diagramm 1: Absorption von $\beta$-Strahlung in Aluminium', size=20)
plt.grid(ls='dotted')

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V253Diagramm1.pdf', format='PDF')
```

```python
#Güte des Fits
chi_squared=np.sum((exp_fit(x1,*popt1)-n_beta_korr)**2/n_beta_korr_err**2)

#Freiheitsgrade
dof=len(n_beta_korr)-2

chi_squared_red=chi_squared/dof

print('Wir erhalten die nachfolgenden Werte für die Güte des Fits:')
print('chi_squared= ' + str(chi_squared))
print('chi_squared_red= ' + str(chi_squared_red))
print()

#Fitwahrscheinlichkeit
prob=round(1-chi2.cdf(chi_squared,dof),2)*100

print('Die Fitwahrscheinlichkeit beträgt: ' + str(prob) + ' %')
```

### Teil 3: Absorption von $\gamma$-Strahlung in Blei

```python
#Kennnummer des Präparats: A AE 8603

#Anzahl Zerfälle
n_gamma=np.array([3950,2623,2052,1566,1094,861,621,513,377,306,225])
n_gamma_err=np.sqrt(n_gamma)

#Torrzeit immer 60s

#Absorberdicke in mm
x2=np.array([0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0])
#fehlerfrei?

#Abzug der Nullmessung unter Berücksichtigung der jeweiligen Torrzeit
n_gamma_korr=n_gamma-(n0_s*60)*np.ones(11)
n_gamma_korr_err=np.sqrt(n_gamma_err**2+(60*n0_s_err*np.ones(11))**2)
```

```python
#linearer Fit
popt2,pcov2=curve_fit(exp_fit,x2,n_gamma_korr,sigma=n_gamma_korr_err)

#Plot der Messdaten
plt.errorbar(x2,n_gamma_korr, yerr=n_gamma_korr_err,linewidth=1,linestyle='',marker='.',color='darkblue',label='Messwerte mit Fehler')
plt.plot(np.linspace(-0.3,5.3,2),exp_fit(np.linspace(-0.3,5.3,2),*popt2),color='darkred',label='Linearer Fit', linewidth=0.8)
plt.xlabel(r'Absorberdicke $ x$  [mm]', size=20)
plt.ylabel(r'Anzahl der Zerfälle', size=20)
plt.yscale('log')
plt.legend(frameon=True,fontsize='large',loc='best', borderpad=1.5, borderaxespad=1.5)
plt.title(r'Diagramm 2: Absorption von $\gamma$-Strahlung in Blei', size=20)
plt.grid(ls='dotted')

#Abspeichern des Bildes im Format "PDF"
plt.tight_layout()
plt.savefig('Diagramme/V253Diagramm2.pdf', format='PDF')
```

```python
#Güte des Fits
chi_squared2=np.sum((exp_fit(x2,*popt2)-n_gamma_korr)**2/n_gamma_korr_err**2)

#Freiheitsgrade
dof2=len(n_gamma_korr)-2

chi_squared2_red=chi_squared2/dof2

print('Wir erhalten die nachfolgenden Werte für die Güte des Fits:')
print('chi_squared= ' + str(chi_squared2))
print('chi_squared_red= ' + str(chi_squared2_red))
print()

#Fitwahrscheinlichkeit
prob2=round(1-chi2.cdf(chi_squared2,dof2),2)*100

print('Die Fitwahrscheinlichkeit beträgt: ' + str(prob2) + ' %')
```

```python
#Schwächungskoeffizient aus Geradensteigung
mu=popt2[1]*1e3
mu_err=(pcov2[1,1])**0.5*1e3
print('Aus dem linearen Fit folgt für den Schwächungskoeffizienten:')
print('mu = (' + str(mu) + ' +/- ' + str(mu_err) + ') 1/m' )

#Massenschwächungskoeffizient
rho_pb=11.34e3 #kg/m^3

mu_rho=mu/rho_pb
mu_rho_err=mu_rho*(mu_err/mu)

print()
print('Für den materialunabhängigen Massenschwächungskoeffizienten mu/rho folgt:')
print('mu/rho = (' + str(mu_rho) + ' +/- ' + str(mu_rho_err) + ') m^2/kg')
print('       = (' + str(mu_rho*10) + '  +/- ' + str(mu_rho_err*10) + ')  cm^2/g')
```
